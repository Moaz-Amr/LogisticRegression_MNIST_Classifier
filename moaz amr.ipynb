```python
import numpy as np
import pandas as pd
# TensorFlow framework
from tensorflow import keras
# Keras algorithms for deep learning and loading datasets
from keras.datasets import mnist
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=mnist.load_data()
data
```

```python
(x_train,y_train),(x_test,y_test)=data
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.axis('off')
plt.show()
```

```python
x_train.shape
# عدد الصور و حجمهم
```

```python
# -1 تعني اني اخلي المودل يحسب الصفوف
x_train_final=x_train.reshape(-1,784)
x_train_final.shape
# 60000, 784)عدد عناصر الفيكتور
```

```python
x_test_final=x_test.reshape(-1,784)
x_test_final.shape
```

```python
# محتاج اعمل نورمليز علشان اخفف الجهد علي المعالج بدل ما يعالج ارقام مابين 255 و 0 هخليها 1 و0
```

```python
# بم ان القيم عندي ابيض  و اسود يعني 255 و 0 لو قسمتهم علي 255 هتتراوح القيم بين ال 0و1
x_test_final=x_test_final/255
x_train_final=x_train_final/255
```

```python
# make model and train it
model=LogisticRegression()
model.fit(x_train_final,y_train)
```

```python
# calculate accuracy of model
y_pred =model.predict(x_test_final)
acc=accuracy_score(y_test,y_pred)
acc
```
